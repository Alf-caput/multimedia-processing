```python
import cv2
import matplotlib.pyplot as plt	

image = cv2.imread('images/coche.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blur = cv2.blur(gray, (5, 5))
plt.imshow(cv2.cvtColor(blur, cv2.COLOR_GRAY2RGB))
plt.axis('off')  # Hide axes
plt.show()
```

```python
edge = cv2.Canny(blur, 30, 200)
contours, _ = cv2.findContours(edge, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(image, contours, -1, (0, 255, 0), 3)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Hide axes
plt.show()
```

```python
image = cv2.imread('images/coche.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blur = cv2.blur(gray, (5, 5))
edge = cv2.Canny(blur, 30, 200)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (15, 15))
# dilate = cv2.dilate(edge, kernel)
close = cv2.morphologyEx(edge, cv2.MORPH_CLOSE, kernel)
contours, _ = cv2.findContours(close, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
for c in contours:
    area = cv2.contourArea(c)
    perimeter = cv2.arcLength(c, True)
    vertices = cv2.approxPolyDP(c, 0.02 * perimeter, True)
    if len(vertices) == 4:
        cv2.drawContours(image, [c], -1, (0, 255, 0), 3)
    # cv2.drawContours(image, [c], -1, (0, 255, 0), 3)
plt.figure()
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
# plt.imshow(cv2.cvtColor(close, cv2.COLOR_BGR2RGB))
# fig, axis = plt.subplots(4, 1) 
# axis[0].imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB))
# axis[1].imshow(cv2.cvtColor(edge, cv2.COLOR_BGR2RGB))
# axis[2].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Hide axes
plt.show()
```

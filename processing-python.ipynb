```python
import cv2
mario = cv2.imread("mario.png")
mario.shape
```

```python
import matplotlib.pyplot as plt

plt.imshow(mario)
plt.show()
```

```python
mario_rgb = cv2.cvtColor(mario, cv2.COLOR_BGR2RGB)
plt.imshow(mario_rgb)
plt.show()
```

```python
mario_gray = cv2.cvtColor(mario, cv2.COLOR_BGR2GRAY)
plt.imshow(mario_gray, cmap="gray")
plt.show()
```

```python
_, mario_bw = cv2.threshold(mario_gray, 127, 255, cv2.THRESH_BINARY)
plt.imshow(mario_bw, cmap="gray")
plt.show()
```

```python
mario_blur = cv2.blur(mario_bw, (16, 16), 0)
plt.imshow(mario_blur, cmap="gray")
plt.show()
```

```python
mario_dilate = cv2.dilate(mario_bw, (16, 16), 0, iterations=10)
plt.imshow(mario_dilate, cmap="gray")
plt.show()
```

```python
mario_dilate = cv2.erode(mario_bw, (16, 16), 0, iterations=10)
plt.imshow(mario_dilate, cmap="gray")
plt.show()
```

```python
mario_open = cv2.morphologyEx(mario_bw, cv2.MORPH_OPEN, (16, 16))
plt.imshow(mario_open, cmap="gray")
plt.show()
```

```python
countours, hierachy = cv2.findContours(mario_bw, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
len(countours)
```
